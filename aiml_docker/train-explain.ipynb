```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(321)
```

```python
f = lambda x: x ** 2 - x + 0.25 

x_obs = np.linspace(0, 1, num=10)
y = f(x_obs) + np.random.normal(0, 0.02, size=len(x_obs))

X = np.array([[x ** 2, x, 1] for x in x_obs])
beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y
beta_hat
```

```python
x_eval = np.linspace(0, 1, num=100)
f_eval = f(x_eval)

f_hat = lambda x: beta_hat[0] * x ** 2 + beta_hat[1] * x + beta_hat[2]
f_hat_eval = f_hat(x_eval)

plt.figure(figsize=(10, 8))
plt.plot(x_eval, f_eval, 'k', label='actual function')
plt.plot(x_obs, y, 'ko', label='simulated observations')
plt.plot(x_eval, f_hat_eval, 'r', label='fitted function')
plt.legend(loc="lower left")
plt.show()
```
