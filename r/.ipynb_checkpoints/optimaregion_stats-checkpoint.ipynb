# Download Statistics of Statistical Software [OptimaRegion](https://cran.r-project.org/web/packages/OptimaRegion/index.html)

Authors: Enrique del Castillo, **Peng Chen**, Adam Meyers, John Hunt, and James Rapkin

```r
# install.packages("cranlogs")

library(cranlogs)
library(tidyverse)
library(glue)
library(scales)
```

```r
df <- cran_downloads("OptimaRegion", from = "2016-01-01", to = Sys.Date()) %>%
    filter(count != 0) %>%
    mutate(cum_dl = cumsum(count))

total_download <- df[["cum_dl"]] %>% tail(1)

df %>%
    ggplot(aes(x = date, y = cum_dl)) +
    geom_line(linewidth = 0.75) + 
    labs(title = glue("Total cumulative downloads to date: {comma(total_download)} times."), 
         x = "", y = "Cumulative downloads") + 
    scale_y_continuous(label = comma) + 
    theme_light()
```
