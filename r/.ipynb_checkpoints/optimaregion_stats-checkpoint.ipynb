```r
install.packages("cranlogs")
```

```r
library(cranlogs)
library(tidyverse)
library(glue)
```

```r
df = cran_downloads("OptimaRegion", from = "2016-01-01", to = Sys.Date()) %>%
    filter(count != 0) %>%
    mutate(cum_dl = cumsum(count))
```

```r
df %>%
    ggplot(aes(x = date, y = cum_dl)) +
    geom_line() + 
    labs(title = "Cumulative Downloads of Statistical Software OptimaRegion ", x = "", y = "")
```

```r
total_download = df[["cum_dl"]] %>% tail(1)
print(glue("Total cumulative download to date: {total_download} times."))
```
